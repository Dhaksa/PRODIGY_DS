```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("/content/exams.csv")
```

```python
dataset.columns
```

```python
avg_math_scores = dataset.groupby('gender')['math score'].mean().reset_index()

# Plotting
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='math score', data=avg_math_scores)
plt.title('Average Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Math Score')
plt.ylim(0, 100)  # Assuming math scores range from 0 to 100
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=dataset, x='math score', kde=True, color='blue', bins=10)
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=dataset, x='gender', palette='viridis')
plt.title('Distribution of Genders')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
